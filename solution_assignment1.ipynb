18. Look up the concepts of *diagonal matrix* and *trace matrix* online and write a small explanation about what they are.


 diagonal matrix in linear algebra means that usually in a squared matrix, all elements outside the diagonal are zeros, and elements of the diagonal are either all ones or has other values.


np.diag() finds us the diagonal elements of a matrix. we can specify the offset as a parameter in this function. 
if the passed array is is v. If v is a 2-D array, the function returns a copy of its k-th diagonal. If v is a 1-D array, the function returns a 2-D array with v on the k-th diagonal.

23. Construct an array of 50 random floats between 0 and 1 and take the average. Moreover;
- Get the standard deviation.
- Get the variance.
- Take the median.
- Find the smallest value.
- Find the largest value.

```python
array23 = np.random.uniform(0,1,50)  
print(np.around(array23, decimals=2))
print("the avg is: ", np.around(array23.mean(), decimals=2)  )
print("the standard deviation. is: ", np.around(array23.std() , decimals=2)  )
print("the variance is: ", np.around(array23.var() , decimals=2) )
print("the smallest value:", np.around(  array23.min(), decimals=2))
print("the largest value is:",  np.around( array23.max(), decimals=2) )
```
